```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import os
import numpy as np
```

```python
def get_address(st_name):
    url = f"http://swopenapi.seoul.go.kr/api/subway/776559696b726b6d3435494b495048/\
            xml/stationInfo/0/7/{st_name}"
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    address = soup.find_all('adresbass')[0].text
    return address
```
